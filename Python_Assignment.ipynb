Kullanıcıdan Celsius sıcaklığını girmesini isteyen (ondalık bir sayı olabilir), girilen sıcaklığı Fahrenheit dereceye çeviren ve sonucu yazdıran kısa bir Python programı yazın.

```python
sayi = input("Lütfen bir sayı giriniz: ")
yeni_sayi= int(sayi)
celsius = yeni_sayi
fahrenheit = celsius * 1.8
f"sıcaklıgı {celsius} olan sayinin fahrenheit {fahrenheit} dir"

```

```python
number = input("Lütfen bir sayı giriniz: ")
number_2 = int(number)
kilometer = number_2
mil = kilometer * 0.62137
f"uzakligi {number_2} olan sayinin mili {mil} dir"
```
