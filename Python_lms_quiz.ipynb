Re-arrange the sentence using the escape sequences from the list using .format() and list indexing methods to get the following output.

Desired output :
I am 40 years old.
    I have two children.
        Data Science is my IT domain.
The string variable to be arranged :
sentence = "I am 40 years old. I have two children. Data Science is my IT domain."
Note that; the second line of the output has one indentation, the third line of the output has two indentations.
(Aşağıdaki çıktıyı almak için .format() ve liste indeksleme yöntemlerini kullanarak listeden kaçış dizilerini kullanarak cümleyi yeniden düzenleyin.

Istenilen çıktı:
Ben 40 yaşındayım.
      İki çocuğum var.
          Veri Bilimi benim BT alanım.
Düzenlenecek dize değişkeni:
cümle = "40 yaşındayım. İki çocuğum var. Veri Bilimi benim BT alanım."
Dikkat; çıktının ikinci satırında bir girinti vardır, çıktının üçüncü satırında iki girinti vardır)

```python
escapes = ["\n\t", ("\t", "\t\t"), ["\n", "\n\t\t"]]

sentence = "I am 40 years old.{} I have two children.{} Data Science is my IT domain.".format(escapes[0], escapes[2][1])

print(sentence)
```
